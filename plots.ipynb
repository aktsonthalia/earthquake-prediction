```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = ['All','Light','Moderate', 'Strong']
A = [91.3, 89.8, 51.0, 9.6]
B = [92.1, 92.7, 61.3, 31.9]
C = [91.7, 91.2, 55.6, 14.7]

width = 0.8
vals = [A, B, C]
n = len(vals)
_X = np.arange(len(X))

p0 = plt.bar(_X - width/2. + 0/float(n)*width, vals[0], 
width=width/float(n), align="edge")   
plt.xticks(_X, X)

p1 = plt.bar(_X - width/2. + 1/float(n)*width, vals[1], 
width=width/float(n), align="edge")   
plt.xticks(_X, X)

p2 = plt.bar(_X - width/2. + 2/float(n)*width, vals[2], 
width=width/float(n), align="edge")   
plt.xticks(_X, X)


plt.legend((p0, p1, p2), ('Precision', 'Recall', 'F1 Score'))
plt.save('plot.png')
plt.show()
```
